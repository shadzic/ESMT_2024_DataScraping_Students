```python
import time
```

```python
#Sample data for Mitte
url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
params = {
    "location": "52.5200,13.4050",  
    "radius": 2000,  
    "type": "restaurant",  
    "key": "AIzaSyDtBfthel4xKUUfTLzSgnppXdafBo1s75M",  
    "pagetoken": "" 
}

all_places = []
Mitte_data = []

while len(all_places) < 60:  
    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()

        all_places.extend(data['results'])

        next_page_token = data.get('next_page_token')
        if next_page_token and len(all_places) < 60:
            params['pagetoken'] = next_page_token
            time.sleep(2)  
        else:
            break
    else:
        print(f"Error: {response.status_code}")
        break

for i, place in enumerate(all_places, start=1):
    name = place.get('name', 'No name provided')
    address = place.get('vicinity', 'No address provided')
    rating = place.get('rating', 'No rating available') 
    user_ratings_total = place.get('user_ratings_total', 'No reviews') 
    
    print(f"{i}. Name: {name}, Address: {address}, Rating: {rating}, Reviews: {user_ratings_total}")

    Mitte_data.append([name, address, rating, user_ratings_total])

df = pd.DataFrame(Mitte_data, columns=['Name', 'Address', 'Rating', 'Reviews']) 
df.to_csv("Mitte_data.csv", index=False) 
```
