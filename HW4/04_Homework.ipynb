```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
df = pd.read_csv("./data/WHO_COVID19_cases.csv")
covid_trio = df[(df.Country == "Germany") | (df.Country == "Italy") | (df.Country == "Japan")]
covid_trio
covid_trio
```

```python
sns.lineplot(x = "Date_reported",
             y = "Cumulative_cases",
             hue = "Country",
             data = covid_trio)

# Title & Labels
plt.title("COVID-19 Cumulative Cases")
plt.xlabel("Date Reported")
plt.ylabel("Cumulative Cases")

# Adjusting Axis Labels
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(20))

# Show results
plt.show()
```

## 2- Make a line plot showing the cumulative_deaths vs cumulative_cases for 3 countries using `plotly` (7.5 points)

```python
import plotly.express as px
import plotly.io as pio
 
pio.renderers.default = "notebook" 
```

```python
fig = px.line(covid_trio, 
    x = "Cumulative_deaths", 
    y = "Cumulative_cases", 
    color = "Country",
    title = "Covid Cases to Covid Deaths Ratio (2020-24)",
    labels = {"Cumulative_deaths": "Cumulative Death", "Cumulative_cases": "Cumulative Cases"})

fig.show(renderer = "browser")
```
